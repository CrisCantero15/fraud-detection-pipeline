# Fase 1: Configuración e Ingesta de Datos

```python
# Instalar e importar las librerías necesarias

!pip install pandas seaborn matplotlib numpy

import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from datetime import datetime
```

```python
# Cargar transacciones.csv y tickets_soporte.json en DataFrames de Pandas

df_transacciones = pd.read_csv("transacciones.csv") 
df_tickets = pd.read_json("tickets_soporte.json")

print("--- Fichero correspondiente a las transacciones realizadas ---\n")
print(df_transacciones.head(), "\n")
print("--- Fichero correspondiente a los tickets de soporte ---\n")
print(df_tickets.head(), "\n")

# Cargar logs_web.txt

registros = []

with open("logs_web.txt", "r", encoding="utf-8") as f:
    for linea in f:
        registros.append([registro.strip() for registro in linea.split("--")]) # Utilizar "--" como separador entre datos dentro de cada línea de registro

df_logs = pd.DataFrame(registros, columns=[
    "timestamp",
    "ip_address",
    "user_id",
    "connection",
    "status_code"
])

# Utilizar expresiones regulares (RegEx) para terminar de obtener todos los datos de cada registro de forma eficiente

df_logs["user_id"] = df_logs["user_id"].str.replace("user:", "", regex=False)

regex = r'"(\w+)"\s+(\S+)\s+(HTTP/\d\.\d)'
df_logs[["method", "endpoint", "http_version"]] = df_logs["connection"].str.extract(regex)
df_logs.drop(columns=["connection"], inplace=True)

print("--- Fichero correspondiente a logs en la web ---\n")
print(df_logs.head(20), "\n")
```

# Fase 2: Diagnóstico y Análisis Exploratorio

```python
# Para cada DataFrame, realiza un diagnóstico para identificar problemas de calidad, como tipos de datos incorrectos, valores nulos, etc.

def diagnostico_calidad(df, nombre_df):
    print(f"\n{'='*60}")
    print(f"DIAGNÓSTICO DEL DATAFRAME: {nombre_df}")
    print(f"{'='*60}")
    
    # Información general
    print("\n--- INFO GENERAL ---")
    df.info()
    
    # Conteo de nulos reales
    print("\n--- NULOS REALES (isnull) ---")
    print(df.isnull().sum())
    
    print("\n--- % DE NULOS ---")
    print((df.isnull().mean() * 100).round(2))
    
    # Valores sospechosos típicos
    print("\n--- VALORES SOSPECHOSOS ---")
    valores_sospechosos = ["-", "null", "NULL", "None", "", " "]
    
    for val in valores_sospechosos:
        conteo = (df == val).sum().sum()
        if conteo > 0:
            print(f"Valor '{val}' encontrado {conteo} veces")
    
    # Detección de tipos incorrectos
    print("\n--- COLUMNAS OBJECT (posibles tipos incorrectos) ---")
    print(df.select_dtypes(include=["object", "string"]).columns.tolist())

diagnostico_calidad(df_transacciones, "Transacciones")
diagnostico_calidad(df_tickets, "Tickets Soporte")
diagnostico_calidad(df_logs, "Logs Web")
```

## 📊 Conclusión del Diagnóstico Inicial

Tras analizar los tres DataFrames, se observan los siguientes puntos clave:

- La columna `metodo_pago` tiene un 24,67% de valores nulos.
- `fecha_transaccion` está como `str` y debería convertirse a `datetime`.
- El resto de columnas no presenta problemas aparentes.

### Tickets Soporte
- 1.762 registros.
- No hay valores nulos.
- `fecha_creacion` también debería convertirse a `datetime`.
- `detalles_ticket` es texto libre (posible análisis posterior).

### Logs Web
- 2.000.000 registros (dataset grande).
- No hay nulos reales, pero existen 599.946 valores `"-"` que probablemente representan datos faltantes.
- Todas las columnas están como `str` y requieren tipado adecuado (`datetime`, `int`, etc.).

---

## 🎯 Resumen General

El principal trabajo de limpieza consistirá en:

- Normalizar valores faltantes (`"-"` → `NaN`).
- Convertir columnas de fechas a `datetime`.
- Ajustar tipos de datos para mejorar calidad y eficiencia.
- Revisar la columna `metodo_pago` por su alto porcentaje de nulos.

Con esto, los DataFrames quedarán listos para la fase de transformación y feature engineering dentro del pipeline ETL.

```python
# Antes de empezar con la fase de limpieza y transformación de datos, se procede a convertir los valores catalogados como sospechosos en valores de tipo NaN

# Valores sospechosos típicos
valores_nulos = ["-", "null", "NULL", "None", "", " "]

# Reemplazar en cada DataFrame
df_transacciones.replace(valores_nulos, np.nan, inplace=True)
df_tickets.replace(valores_nulos, np.nan, inplace=True)
df_logs.replace(valores_nulos, np.nan, inplace=True)

print("Valores nulos normalizados a NaN")

print(df_transacciones.isna().sum())
print(df_tickets.isna().sum())
print(df_logs.isna().sum())
```

# Fase 3: Limpieza y Transformación

```python
## Para df_transacciones:
## - Asegúrate de que fecha_transaccion sea del tipo datetime.
## - Estandariza la columna estado (ej. "approved", "APPROVED", "success" -> "APROBADA").
## - Gestiona los valores nulos en metodo_pago.

print("\n--- Tipo de dato de la columna 'fecha_transaccion' del DataFrame de transacciones ---\n")
df_transacciones["fecha_transaccion"] = pd.to_datetime(df_transacciones["fecha_transaccion"], errors="coerce")
print(df_transacciones["fecha_transaccion"].dtype)
print(f"Nulos en la columna: {df_transacciones['fecha_transaccion'].isna().sum()}")

print("\n--- Valores existentes en la columna 'estado' del DataFrame de transacciones ---\n")
df_transacciones.loc[df_transacciones["estado"].isin(["approved", "APPROVED", "success"]), "estado"] = "APROBADA"
df_transacciones.loc[df_transacciones["estado"].isin(["denied", "failed"]), "estado"] = "RECHAZADA"
df_transacciones.loc[df_transacciones["estado"].isin(["pending"]), "estado"] = "PENDIENTE"
print(df_transacciones["estado"].value_counts(dropna=False))

print("\n--- Normalización de valores nulos en la columna 'metodo_pago' del DataFrame de transacciones ---\n")
df_transacciones["metodo_pago"] = df_transacciones["metodo_pago"].str.upper()
df_transacciones["metodo_pago"] = df_transacciones["metodo_pago"].fillna("OTHER")
print(df_transacciones["metodo_pago"].isna().sum())
print(df_transacciones["metodo_pago"].value_counts(dropna=False).head(10))

## Para df_tickets:
## - Normaliza la columna anidada detalles_ticket.
## - Crea una columna booleana es_reporte_fraude basada en el texto del ticket.

print("\n--- Conversión a tipo Datetime de la columna 'fecha_creacion' en el DataFrame de tickets ---\n")
df_tickets["fecha_creacion"] = pd.to_datetime(df_tickets["fecha_creacion"], errors="coerce")

print("\n--- Normalización de la columna anidada 'detalles_ticket' del DataFrame de tickets ---\n")
detalles_json = pd.json_normalize(df_tickets["detalles_ticket"])
df_tickets = pd.concat([df_tickets, detalles_json], axis=1)
df_tickets.drop(columns=["detalles_ticket"], inplace=True)
df_tickets.info()

print("\n--- Crear columna booleana 'es_reporte_fraude' del DataFrame de tickets ---\n")
palabras_fraude = ["fraude", "fraud", "robo", "hack", "phishing"]
df_tickets["descripcion"] = df_tickets["descripcion"].str.lower()
df_tickets["es_reporte_fraude"] = False
df_tickets.loc[
    df_tickets["descripcion"].str.contains(
        "|".join(palabras_fraude),
        case=False,
        na=False
    ), 
    "es_reporte_fraude"    
] = True
print(df_tickets["es_reporte_fraude"].value_counts(normalize=True, dropna=False) * 100)

## Para df_logs:
## - Limpia y convierte los tipos de datos de las columnas parseadas (ej. user_id a numérico, status_code a entero).

print("\n--- Convertimos los tipos de datos correctamente del DataFrame de logs ---\n")

df_logs["timestamp"] = pd.to_datetime(df_logs["timestamp"], errors="coerce")
df_logs["user_id"] = pd.to_numeric(df_logs["user_id"], errors="coerce")
df_logs["status_code"] = pd.to_numeric(df_logs["status_code"], errors="coerce").astype('Int64')

df_logs.info() # Verificamos los tipos de datos que tenemos ahora
```

Aunque la tasa de **es_reporte_fraude = True (71.6%)** es correcta desde el punto de vista matemático (1262 de 1762 tickets), este resultado requiere un análisis más profundo debido a su magnitud.

En primer lugar, se trata de una proporción anormalmente elevada si se compara con escenarios reales, donde los casos de fraude suelen representar un porcentaje mucho menor del total. Esto sugiere que el dataset podría no ser representativo de la realidad general, sino estar condicionado por su origen. Por ejemplo, es posible que los datos provengan de un sistema especializado en la gestión de incidencias relacionadas con fraude, lo que explicaría la sobrerrepresentación de este tipo de casos.

Por otro lado, también cabe considerar la posibilidad de que exista algún tipo de sesgo en la recogida o en el etiquetado de los datos, ya sea por criterios amplios a la hora de clasificar un ticket como fraude o por una selección previa de los casos incluidos en el dataset.

Desde el punto de vista analítico, este desequilibrio tiene implicaciones importantes. En caso de utilizar estos datos para entrenar modelos de machine learning, la alta proporción de una clase frente a la otra podría provocar que el modelo aprenda patrones sesgados, favoreciendo la clase mayoritaria. Esto haría necesario aplicar técnicas de balanceo o utilizar métricas de evaluación adecuadas como precision, recall o F1-score, en lugar de basarse únicamente en la exactitud (accuracy).

En conclusión, aunque el cálculo es correcto, el valor obtenido debe ser interpretado con cautela, analizando el contexto y las posibles limitaciones del dataset.

# Fase 4: Enriquecimiento de Datos (Feature Engineering)

```python
## Unión de Datos: Une (merge) el DataFrame de transacciones con el de tickets para identificar qué transacciones fueron posteriormente reportadas como fraude.
## Crea una nueva columna hora_transaccion en el DataFrame de transacciones.

print(df_tickets["descripcion"].head(20))

# Observamos que en algunos tickets aparece el número de transacción, así que siguiente paso: 
# Crear la columna id_transaccion en df_tickets a partir de la descripcion para poder mergear ambos DataFrames

df_tickets["id_transaccion"] = df_tickets["descripcion"].str.extract(
    r'([0-9a-fA-F]{8}-[0-9a-fA-F]{4}-[0-9a-fA-F]{4}-[0-9a-fA-F]{4}-[0-9a-fA-F]{12})',
    expand=False
)
df_tickets["id_transaccion"] = df_tickets["id_transaccion"].fillna("SIN_TRANSACCION")

print(f"\nValores de la nueva columna id_transaccion:\n {df_tickets['id_transaccion'].value_counts(dropna=False).head(10)}\n")

df_transacciones_tickets = pd.merge(
    df_transacciones,
    df_tickets,
    on="id_transaccion",
    how="left"
)

print("\n--- Nuevo DataFrame resultante de mergear transacciones y tickets ---\n")

df_transacciones_tickets.info()

print("\n--- Crear nueva columna hora_transaccion en el DataFrame de transacciones ---\n")
df_transacciones["hora_transaccion"] = df_transacciones["fecha_transaccion"].dt.hour
print(f"Valores de hora_transaccion: {df_transacciones['hora_transaccion'].value_counts(dropna=False).head(10)}")
```

# Fase 5: Almacenamiento Optimizado en el Data Lake

```python
# Crear estructura Data Lake
os.makedirs("ecommerce_data_lake/raw", exist_ok=True)
os.makedirs("ecommerce_data_lake/processed", exist_ok=True)

# Transacciones
df_transacciones.to_parquet(
    "ecommerce_data_lake/processed/transacciones.parquet",
    index=False
)

# Logs
df_logs.to_parquet(
    "ecommerce_data_lake/processed/logs.parquet",
    index=False
)

# Tickets
df_tickets.to_parquet(
    "ecommerce_data_lake/processed/tickets.parquet",
    index=False
)

# Transacciones + Tickets
df_transacciones_tickets.to_parquet(
    "ecommerce_data_lake/processed/transacciones_tickets.parquet",
    index=False
)

print("Data Lake creado y procesado correctamente.")
```

# Fase 6: Conclusiones y Justificación

### ¿Por qué fue el parsing de los logs web (logs_web.txt) el paso de ingesta más complejo?

El parsing de los logs web (logs_web.txt) representó el paso de ingesta más complejo debido a que se trataba de datos semi-estructurados, a diferencia de otros formatos más estructurados como CSV o JSON. Los logs web contenían múltiples atributos codificados dentro de una misma línea de texto, por lo que fue necesario aplicar técnicas de procesamiento de texto y extracción de patrones para transformar la información en un formato tabular adecuado para su análisis mediante Pandas.

En concreto, fue necesario identificar y separar campos como timestamp, dirección IP, identificador de usuario, método HTTP, endpoint, versión del protocolo y código de respuesta utilizando delimitadores y expresiones regulares. Este proceso implica una mayor complejidad lógica en comparación con la carga directa de datasets estructurados, ya que el parsing debe ser capaz de manejar variaciones en el formato del texto y posibles inconsistencias en los registros. De hecho, esta dificultad se volvió a ver a la hora de extraer el ID de transacción desde la descripción de los diferetes tickets.

Adicionalmente, fue necesario realizar un proceso de limpieza posterior para tratar representaciones no estándar de datos, como caracteres especiales o valores simbólicos como -, que debieron ser convertidos a valores nulos para mantener la consistencia del dataset.

Desde el punto de vista del pipeline de datos, este paso es fundamental en arquitecturas ETL modernas, ya que garantiza que los datos puedan ser posteriormente utilizados en tareas de análisis exploratorio o modelado predictivo.

### ¿Qué ventaja clave ofrece tener un Data Lake unificado con transacciones, logs y tickets para la detección de fraude, en comparación con tenerlos en sistemas separados?

La principal ventaja de un Data Lake unificado radica en la capacidad de realizar un análisis exploratorio entre múltiples fuentes de datos, permitiendo obtener una visión general del comportamiento del usuario y mejorar la detección de actividades fraudulentas. En sistemas tradicionales, la información de transacciones, actividad web y tickets de soporte se encuentra distribuida en diferentes sistemas, lo que limita la capacidad de detectar patrones complejos de fraude que requieren la combinación de múltiples señales de comportamiento.

En cambio, un Data Lake permite integrar estos datos en un mismo entorno, facilitando la construcción de features más ricas para modelos de ML y análisis de comportamiento. Por ejemplo, es posible correlacionar:

* Eventos de navegación detectados en los logs web.
* Operaciones financieras registradas en las transacciones.
* Reportes de incidencias o fraude detectados en tickets de soporte.

Desde la perspectiva de la Inteligencia Artificial y el Análisis de Datos, esta unificación permite mejorar la capacidad predictiva de los modelos al proporcionar contexto adicional, lo que resulta especialmente útil en sistemas de detección de fraude basados en aprendizaje supervisado o análisis de anomalías. Desde el punto de vista del negocio, un Data Lake unificado mejora la toma de decisiones basada en datos y permite la implementación de sistemas de monitorización y alerta temprana frente a comportamientos sospechosos.
